```python
import pandas as pd
df= pd.read_csv("C:/Users/user/Desktop/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For cleaner visuals
sns.set(style="whitegrid")
```

```python
#Countplot of churn
sns.countplot(x='Churn', data=df, palette='Set2') # Creates bar chart. X axis- churn values(yes/no) Y axis- churn count
plt.title('Churn Distribution')    # Gives title
plt.show()                         # To display

# Percentage of Churned vs Retained
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
# Insight Churn Distribution
73.4% people have not churned.
26.5% people have churned.
This shows a moderately high churn risk, so retention strategies are needed.
```

```python
sns.countplot(x='gender', hue='Churn', data=df, palette='Set1') #X axis has male & female gender, churned or not , set color
plt.title('Gender-wise Churn')              # Add title
plt.show()                                  # Display
```

```python
# Insight Gender- wise churn
Both male and female customers predominantly do not churn. However, a slightly higher number of female customers churn compared to 
male customers.
Gender does not significantly affect churn in this dataset
```

```python
sns.countplot(x='Contract', hue='Churn', data=df, palette='Pastel2') 
plt.title('Churn by Contract Type')
plt.show()
```

```python
# Insight Churn by Contract type
EDA reveals that customers on month to month contracts have the highest churn , suggesting a need for
long term engangement stratergies.
Customers with 1 year or 2 year contracts are more loyal
Business should perform long term plans to reduce churn.
```

```python
sns.countplot(x='InternetService', hue='Churn', data=df, palette='Set3')
plt.title('Churn by Internet Service Type')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette='coolwarm')
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
# Insight: Monthly Charges vs Churn
- Customers who churn generally pay *higher monthly charges*.
- This suggests that *pricing* may be a reason for churn.
- Business should explore *discounts, **bundled offers, or **value-added services* to retain these users.
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30, palette='cool')
plt.title('Tenure Distribution by Churn Status')
plt.show()
```

```python
# Insight: Tenure Distribution
- Most customers who churn have a *tenure between 012 months*.
- Longer-tenure customers are more likely to stay.
- Suggests that *early churn* is a key challenge  strong onboarding & early engagement is crucial.
```
